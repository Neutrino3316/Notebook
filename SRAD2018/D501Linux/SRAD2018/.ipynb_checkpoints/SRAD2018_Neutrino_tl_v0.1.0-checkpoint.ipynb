# 读取函数

一个时间段里面有61张图片。我们根据3张图片之间的差值预测最新的一个差值。例如我们选0号、5号、10号这三张图片，并且计算出差值作为输入，而输出是10号图片和15号图片的差值

由于雷达的扫描有时候会出现一些问题，导致图片缺失了一部分，我们通过图片像素点的值的总和来判断图片是否有突变。如果存在突变（即很有可能是图片出现了缺失），我们将舍弃这组数据。另外，如果图片像素点的和太小（即很有可能是云层很少，只有一个角落有云，其他都是没有云的缺省值），我们也会舍弃这组数据。

所以输入输出都是图片的插值。

输入最后都会除100再输入神经元网络。而网络的输出会乘100再和最新一个扫描图片相加，得出的就是下一张预测的图片了。

```python
def read_by_RAD_id(RAD_id, step_size):
    
    for k in range(1, 60-4*step_size):
        
        sample_1_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 1 * step_size))).astype(np.int8)
        if sample_1_matrix.ravel().sum() < 251001 * 2:
#             print('skip bad data')
            break
    
        sample_1_l_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 1 * step_size - 1))).astype(np.int8)
        sample_1_r_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 1 * step_size + 1))).astype(np.int8)
        if abs(2*sample_1_matrix.ravel().sum() - (sample_1_l_matrix.ravel().sum()+sample_1_l_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue
    
        sample_0_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 0 * step_size))).astype(np.int8)
        sample_0_l_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 0 * step_size - 1))).astype(np.int8)
        sample_0_r_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 0 * step_size + 1))).astype(np.int8)
        if abs(2*sample_0_matrix.ravel().sum() - (sample_0_l_matrix.ravel().sum()+sample_0_l_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue
    
        sample_2_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 2 * step_size))).astype(np.int8)
        sample_2_l_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 2 * step_size - 1))).astype(np.int8)
        sample_2_r_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 2 * step_size + 1))).astype(np.int8)
        if abs(2*sample_2_matrix.ravel().sum() - (sample_2_l_matrix.ravel().sum()+sample_2_l_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue
        if abs(2*sample_1_matrix.ravel().sum() - (sample_0_matrix.ravel().sum()+sample_2_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue
    
        sample_3_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 3 * step_size))).astype(np.int8)
        sample_3_l_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 3 * step_size - 1))).astype(np.int8)
        sample_3_r_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 3 * step_size + 1))).astype(np.int8)
        if abs(2*sample_3_matrix.ravel().sum() - (sample_3_l_matrix.ravel().sum()+sample_3_l_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue
    
        sample_4_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 4 * step_size))).astype(np.int8)
        sample_4_l_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 4 * step_size - 1))).astype(np.int8)
        sample_4_r_matrix = np.array(PIL.Image.open("/home/aisalon/HZZ/Neutrino/SRAD2018/SRAD2018_train_data/%s/%s_%03d.png" % (RAD_id, RAD_id, k + 4 * step_size + 1))).astype(np.int8)
        if abs(2*sample_4_matrix.ravel().sum() - (sample_4_l_matrix.ravel().sum()+sample_4_l_matrix.ravel().sum())) > 251001 * 3:
#             print('skip bad data')
            continue

        x_0_matrix = sample_3_matrix - sample_2_matrix
        x_1_matrix = sample_2_matrix - sample_1_matrix
        x_2_matrix = sample_1_matrix - sample_0_matrix
        x_3_matrix = (sample_3_matrix - sample_1_matrix) / 2
        x_4_matrix = (sample_2_matrix - sample_0_matrix) / 2
        x_5_matrix = (sample_3_matrix - sample_2_matrix - sample_1_matrix + sample_0_matrix) / 2
        x_matrix = np.empty((6, 501, 501))
        x_matrix[0] = x_0_matrix
        x_matrix[1] = x_1_matrix
        x_matrix[2] = x_2_matrix
        x_matrix[3] = x_3_matrix
        x_matrix[4] = x_4_matrix
        x_matrix[5] = x_5_matrix
        y_matrix = sample_4_matrix - sample_3_matrix
            
        x_matrix = np.stack(x_matrix, axis=-1)
        x_matrix = x_matrix / 100
        y_matrix = y_matrix.reshape(501, 501, 1)
        y_matrix = y_matrix / 100
        return ([x_matrix], [y_matrix])
    
    return (None, None)
```

# 损失函数

损失函数采用了平方误差。插值的取值大概在$[-50,50]$之间，每个像素点的权重设为像素点的绝对值加上10。这样可以使得模型更加关注云层变化比较大的地方，而更少关注云层没有变化的地方。

```python
def my_loss(y_pred, y_true):
    w = tf.add(tf.abs(y_true), tf.constant(10.0))
    return tf.losses.mean_squared_error(y_true, y_pred, weights=w)
```

# 模型结构

先卷积一个卷积核大小为1的卷积层，把通道的数量扩大为128，然后就有四层的卷积，最后一层的卷积核大小是1，而且会将通道数量化会1.

具体结构可以参考下面的代码

```python
kernel_size = 5
tf.reset_default_graph()
tflearn.init_graph(gpu_memory_fraction=0.8, soft_placement=True)
net = tflearn.input_data(shape=[None, 501, 501, 6])
net = tflearn.layers.conv.conv_2d(net, 128, 1, activation='tanh', weights_init=tflearn.initializations.xavier())
net = tflearn.layers.batch_normalization(net)
net = tflearn.layers.dropout(net, 0.8)
net = tflearn.layers.conv.conv_2d(net, 128, kernel_size, activation='elu', weights_init=tflearn.initializations.xavier())
net = tflearn.layers.batch_normalization(net)
net = tflearn.layers.dropout(net, 0.8)
for i in range(3):
    net = tflearn.layers.conv_2d(net, 64, kernel_size, activation='elu', weights_init=tflearn.initializations.xavier())
    net = tflearn.layers.batch_normalization(net)
    net = tflearn.layers.dropout(net, 0.8)
net = tflearn.layers.conv.conv_2d(net, 1, kernel_size, activation='tanh', weights_init=tflearn.initializations.zeros(), bias=False)
optimizer = tflearn.optimizers.adam(learning_rate=0.0001)
net = tflearn.regression(net, optimizer=optimizer, loss=my_loss)
model = tflearn.DNN(net, tensorboard_verbose=3, tensorboard_dir='/home/aisalon/HZZ/tflearn_logs')
# print(time.strftime("build time: %Y-%m-%d %H:%M:%S", time.localtime()))
```

# 训练模型

因为图片很大，所以我们每次只会读取一个时间段来作为一个batch，一个时间段里面的样本数量不确定，可能由于数据不好而全部舍弃、跳过这个时间段。

```python
fitted_batch_num = 0
step_size = 5
v_scalar = 50
start_time = time.time()
loss_decay = []
# print(time.strftime("start time: %Y-%m-%d %H:%M:%S", time.localtime()))
for epoch in range(3):
    for counter, RAD_id in enumerate(RAD_id_list[0:12000]):
        temp_x, temp_y = read_by_RAD_id(RAD_id, step_size)
        if temp_x == None:
#             print('skip bad data')
            continue
        if counter % 20 == 0 and epoch % 30 == 0:
#             clear_output()
            print('↓↓↓counter =', counter, '↓↓↓time =', time.time() - start_time)
            temp_y_true = (temp_y[0] * 100).astype(np.int8).reshape((501,501))
            temp_y_pred = (model.predict([temp_x[0]])[0] * 100).astype(np.int8).reshape((501,501))
            plt.imshow(temp_y_true, cmap=cm.bwr, vmin=-v_scalar, vmax=v_scalar)
            plt.show()
            print('↓↓↓下面的是训练过程中模型的输出。上面这张是对应的真实值↑↑↑')
            plt.imshow(temp_y_pred, cmap=cm.bwr, vmin=-v_scalar, vmax=v_scalar)
            plt.show()
            train_loss = tflearn.mean_square(temp_y_true.ravel().astype(np.int32), temp_y_pred.ravel().astype(np.int32))
#             train_loss = my_loss(temp_y_pred, temp_y_true)
            with tf.Session():
                print('↑↑↑counter =', counter, '↑↑↑loss =', train_loss.eval(), '↑↑↑fitted_batch_num =', fitted_batch_num)
        temp_loss_decay = model.fit_batch(temp_x, temp_y)
        fitted_batch_num += 1
        loss_decay.append(temp_loss_decay)
#         if fitted_batch_num % 50 == 0:
#             plt.plot(loss_decay[-50:])
#             plt.show()
        if temp_loss_decay < 0.005 or len(loss_decay) > 100:
            print('loss is small enough')
            break
    print(fitted_batch_num, time.time() - start_time)
#     print(time.strftime("now: %Y-%m-%d %H:%M:%S", time.localtime()))
    print('↑↑↑epoch=%d↑↑↑' % epoch)
# print(time.strftime("end time: %Y-%m-%d %H:%M:%S", time.localtime()))
plt.plot(loss_decay[-50:])
print('↑↑↑损失函数最后下降到一个比较低的值然后就不下降了↑↑↑')
plt.show()
```
