```python
hash_table_train = {}
start_time = time.time()
for counter, RAD_id in enumerate(RAD_id_train_list):
    if counter % 1000 == 0:
        print(counter, time.time()-start_time)
    for i in range(61):
        with open("C:\\Users\\Administrator\\Downloads\\Downloads\\SRAD2018_train\%s\%s_%03d.png" % (RAD_id, RAD_id, i), 'rb') as afile:
            buf = afile.read()
            hasher = hashlib.md5()
            hasher.update(buf)
        if hasher.hexdigest() in hash_table_train:
            hash_table_train[hasher.hexdigest()].append([RAD_id, i])
        else:
            hash_table_train[hasher.hexdigest()] = [[RAD_id, i]]
            
hash_table_test_1 = {}
start_time = time.time()
for counter, RAD_id in enumerate(RAD_id_test_1_list):
    if counter % 1000 == 0:
        print(counter, time.time()-start_time)
    for i in range(31):
        with open("C:\\Users\\Administrator\\Downloads\\Downloads\\SRAD2018_Test_1\%s\%s_%03d.png" % (RAD_id, RAD_id, i), 'rb') as afile:
            buf = afile.read()
            hasher = hashlib.md5()
            hasher.update(buf)
        if hasher.hexdigest() in hash_table_test_1:
            hash_table_test_1[hasher.hexdigest()].append([RAD_id, i])
        else:
            hash_table_test_1[hasher.hexdigest()] = [[RAD_id, i]]
            
hash_table_test_2 = {}
start_time = time.time()
for counter, RAD_id in enumerate(RAD_id_test_2_list):
    if counter % 1000 == 0:
        print(counter, time.time()-start_time)
    for i in range(31):
        with open("C:\\Users\\Administrator\\Downloads\\Downloads\\[20180919] SRAD2018_Test_2\%s\%s_%03d.png" % (RAD_id, RAD_id, i), 'rb') as afile:
            buf = afile.read()
            hasher = hashlib.md5()
            hasher.update(buf)
        if hasher.hexdigest() in hash_table_test_2:
            hash_table_test_2[hasher.hexdigest()].append([RAD_id, i])
        else:
            hash_table_test_2[hasher.hexdigest()] = [[RAD_id, i]]
```
